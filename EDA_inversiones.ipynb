```
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('inversiones.csv')
df.head()
```

## Auditoría de datos

```
df.info()
df.describe()
df.isnull().sum()
```

## Limpieza

```
df = df.drop_duplicates()
df = df.fillna(0)
```

## Feature Engineering

```
cols_pais = [c for c in df.columns if '_pais' in c]
cols_ext = [c for c in df.columns if '_exterior' in c]

df['total_pais'] = df[cols_pais].sum(axis=1)
df['total_exterior'] = df[cols_ext].sum(axis=1)
df['total_inversion'] = df['total_pais'] + df['total_exterior']
```

## Análisis

```
df[['total_pais','total_exterior']].sum()
```

```
df.sum(numeric_only=True).sort_values(ascending=False).head(10)
```

## Visualizaciones

```
df[['total_pais','total_exterior']].sum().plot(kind='bar')
plt.title('Inversiones País vs Exterior')
plt.show()
```

```
df.sum(numeric_only=True).sort_values(ascending=False).head(10).plot(kind='bar')
plt.title('Top inversiones')
plt.show()
```

## Conclusiones
Alta concentración en inversiones locales y baja diversificación.
